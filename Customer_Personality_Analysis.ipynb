```python
#number of days a customer:
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
dates=[]
for i in df['Dt_Customer']:
  i=i.date()
  dates.append(i)
print('newest customer enrolled dates',max(dates))
print('oldest customer enrolled dates',min(dates))
```

```python
#creating a feature based on lastet dates:
days=[]
d=max(dates)
for i in dates:
  delta=d-i
  days.append(delta)
df['customer']=days
df['customer']=df['customer'].apply(lambda x:x.days)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,accuracy_score
X_0=df.copy()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_0)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values to visualize the Elbow Method
plt.plot(range(1, 11), wcss, '-o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
y_pred = kmeans.fit_predict(X_0)

X_0['Cluster']=y_pred+1

# Plot the count of data points in each cluster
plt.figure(figsize=(10, 6))
sns.countplot(x=X_0['Cluster'], palette='viridis')
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Count of Data Points in Each Cluster')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Exp', y='Income', hue='Cluster', data=X_0, palette='viridis')
plt.xlabel('Experience')
plt.ylabel('Income')
plt.title('KMeans Clustering (2 Clusters)')
plt.show()
```

```python
X_1=df.copy()
X_1.head()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

# Apply KernelPCA with RBF kernel
pca =PCA(n_components=3)
pca.fit(X_1)

pca_ds=pd.DataFrame(pca.transform(X_1),columns=['PC1','PC2','PC3'])
pca_ds.describe().T
```

```python
x=pca_ds['PC1']
y=pca_ds['PC2']
z=pca_ds['PC3']

fig= plt.figure(figsize=(10,8))

ax=fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c='maroon', marker="o")

ax.set_title("A 30 Projection of Data In The Reduced Dimension")

plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(pca_ds)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method
plt.figure(figsize=(16, 8))
plt.plot(range(1, 11), wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-cluster Sum of Squares)')
plt.xticks(np.arange(1, 11, 1))
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns

# Initialize Agglomerative Clustering with 2 clusters
AC = AgglomerativeClustering(n_clusters=2)

# Fit model and predict clusters
yhat_AC = AC.fit_predict(pca_ds)

# Assign clusters to the original dataframe
pca_ds["Clusters"] = yhat_AC

# Adding the Clusters feature to the original dataframe X_1
X_1["Cluster_Agglo"] = yhat_AC + 1

# Visualize clusters using seaborn
sns.scatterplot(data=X_1, x='Exp', y='Income', hue='Cluster_Agglo')
```
